```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
# Load Dataset

df = pd.read_csv("bmw.csv")

# Display first 5 rows
df.head()
```

```python
# Check all values
print(df.isna())
```

```python
# Statistical summary
df.describe(include='all')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=20)
plt.title("Distribution of Car Prices")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=20)
plt.title("Distribution of Car Prices")
plt.show()
```

```python
model_revenue = df.groupby('model')['price'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
model_revenue.plot(kind='bar')
plt.title("Total Revenue by Model")
plt.ylabel("Total Price")
plt.show()

model_revenue
```

```python
plt.figure(figsize=(8,5))                             #Mileage Vs Price
sns.scatterplot(x='mileage', y='price', data=df)
plt.title("Mileage vs Price")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))                           #Year Vs Price
sns.scatterplot(x='year', y='price', data=df)
plt.title("Year vs Price")
plt.xlabel("Year")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['year', 'price', 'mileage', 'tax', 'mpg', 'engineSize']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of BMW Features")
plt.show()
```

```python
year_avg_price = df.groupby('year')['price'].mean()

plt.figure(figsize=(10,5))
year_avg_price.plot(marker='o')
plt.title("Average BMW Price by Year")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.show()
```
